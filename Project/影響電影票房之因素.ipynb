# 一、研究動機
電影對於消費者是相對平價且消費得起的娛樂性產品，近年以來有許多電影院如雨後春筍般的開幕營運，截至行政院文化部106年電影片映統計台灣共有111家電影院、784廳屬於營業中的電影院，而數量也將隨著近期新開幕的電影院而成長(如:桃園統領威秀影城、花蓮新天堂樂園威秀影城)。
                                                                     由此可知國人對於電影的需求是逐年上升，以至於票房破億的電影不在少數，電影產業的熱絡更凸顯台灣人對於電影的熱愛， 因此本組想要探討哪種類型、時期的電影會吸引較多消費者及業者，且可以訂定哪些策略。另外由於網路普及，youtube使用人數逐年上升，也想知道在網路影音平台上的預告片的點閱率是否對電影票房有所影響。

# 五、結論與政策建議 
**結論:**

近年國人對於電影的需求是逐年上升，以至於票房破億的電影不在少數，電影產業的熱絡更凸顯台灣人對於電影的熱愛，為探討影響電影票房的因素，我們以電影票房為應變數，電影上映天數、電影上映院數、寒暑假期上映電影、電影製作國家、電影類別、Youtube上電影官方中文預告觀看次數等為自變數，做OLS簡單回歸分析，由實證結果我們得知，上述自變數僅寒暑假上映電影不會對電影票房有顯著影響，其他則都有顯著影響，R-squared=0.4596也顯示本組所蒐集準備的資料對電影票房分析是有45.96%的解釋能力。

**政策建議:**

這次實證結果中我們發現Youtube上電影官方中文預告觀看次數以及電影類別等會影響電影票房的高低，由此兩點提出政策建議，如果電影公司可以把普遍大眾都較喜歡的類別電影，呈現在同一部電影之中，並多利用網路社群媒體行銷，透過臉書、LINE 等各種社群媒體推播 YouTube 上的電影預告以提升點擊率，及利用最近韓流旋風的趨勢，想必有機會提升電影票房收益。
