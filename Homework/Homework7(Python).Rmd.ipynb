```python
金額 = int(input("輸入金額"))
年收益率 = int(input("輸入收益率"))
經過月數 = int(input("輸日經過月數"))
print ("經過月數","金額")
for i in range(1, 經過月數+1):
    存款 = 金額 * 年收益率/12
    print (i,int(存款))
```
